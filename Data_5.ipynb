```python
# Ques-1
import numpy as np
a=np.random.randint(1,90,25)
x=a.reshape(5,5)
print(x)
print("\n")
# mean
print(np.mean(a))
print(np.mean(x))
print(np.mean(x,axis=0))#columnwise
print(np.mean(x,axis=1))#row-wise
print("\n")
# sum
b=np.sum(x,axis=0)
print(b)
c=np.sum(x,axis=1)
print(c)
print("\n")
# sort
d=np.sort(x,axis=1)
print(d)
print("\n")
e=np.sort(x,axis=0)
print(e)
print("\n")
#replace
for i in range(0,25):
  if(a[i]%2==0):
    a[i]=-1
y=a.reshape(5,5)
print(y)
#count
count_col=np.sum(np.where(y<0,1,0),axis=0)
print(count_col)


```

```python
# Ques-2
import numpy as np
a=np.random.randint(20,80,25)
b=np.random.randint(20,80,25)

stock=a.reshape(5,5)
demand=b.reshape(5,5)
# remaining stock
rstock=stock-demand
print(rstock)
# total shortage
totalShortage = np.sum(np.where(rstock<0,-1*rstock,0),axis = 0)
print("\nTotal Shortage : \n",totalShortage)
# replace zero
mr=np.where(rstock<0,0,rstock)
print(mr)
# zero stock
zero=np.sum(np.where(mr==0,1,0),axis=0)
print(zero)
# row sort
mrp=np.sort(mr,axis=1)
print(mrp)
```
